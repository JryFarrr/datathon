```python
import os
import torch
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import json
import re
from pathlib import Path
from typing import List, Dict, Tuple, Optional
import warnings
warnings.filterwarnings('ignore')

# PDF processing imports
import PyPDF2
import pdfplumber
import fitz  # PyMuPDF
from pdfminer.high_level import extract_text
from pdfminer.layout import LAParams
import nltk
import spacy
from langdetect import detect
import textstat

# Transformers imports
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    T5ForConditionalGeneration,
    T5Tokenizer,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq,
    pipeline
)

from datasets import Dataset

# Download NLTK data
try:
    nltk.download('punkt', quiet=True)
    nltk.download('stopwords', quiet=True)
    print("✅ NLTK data downloaded")
except:
    print("⚠️ NLTK download failed")

# =============================================================================
# 3. PDF PROCESSING CLASS
# =============================================================================

class PDFFinancialExtractor:
    """
    Class untuk ekstraksi dan preprocessing laporan keuangan PDF
    """

    def __init__(self):
        self.supported_formats = ['.pdf', '.PDF']
        self.financial_keywords = [
            'laporan keuangan', 'financial statement', 'pendapatan', 'revenue',
            'laba', 'profit', 'rugi', 'loss', 'aset', 'asset', 'liabilitas',
            'liability', 'ekuitas', 'equity', 'kas', 'cash', 'piutang',
            'receivable', 'persediaan', 'inventory', 'penjualan', 'sales',
            'beban', 'expense', 'biaya', 'cost', 'margin', 'ebitda',
            'neraca', 'balance sheet', 'laporan laba rugi', 'income statement',
            'arus kas', 'cash flow', 'triliun', 'miliar', 'juta', 'ribu'
        ]

        # Load Indonesian language model jika ada
        try:
            self.nlp = spacy.load("id_core_news_sm")
            print("✅ Indonesian spacy model loaded")
        except:
            try:
                self.nlp = spacy.load("en_core_web_sm")
                print("✅ English spacy model loaded")
            except:
                self.nlp = None
                print("⚠️ No spacy model available")

    def extract_text_from_pdf(self, pdf_path: str, method: str = 'pdfplumber') -> Dict:
        """
        Ekstraksi teks dari PDF dengan berbagai metode

        Args:
            pdf_path: Path ke file PDF
            method: Metode ekstraksi ('pdfplumber', 'pymupdf', 'pdfminer')

        Returns:
            Dictionary dengan teks per halaman dan metadata
        """
        try:
            if method == 'pdfplumber':
                return self._extract_with_pdfplumber(pdf_path)
            elif method == 'pymupdf':
                return self._extract_with_pymupdf(pdf_path)
            elif method == 'pdfminer':
                return self._extract_with_pdfminer(pdf_path)
            else:
                raise ValueError(f"Unsupported method: {method}")

        except Exception as e:
            print(f"❌ Error extracting {pdf_path}: {e}")
            return {"pages": [], "metadata": {}, "tables": []}

    def _extract_with_pdfplumber(self, pdf_path: str) -> Dict:
        """Ekstraksi menggunakan pdfplumber (terbaik untuk tabel)"""
        result = {"pages": [], "metadata": {}, "tables": []}

        with pdfplumber.open(pdf_path) as pdf:
            result["metadata"] = {
                "total_pages": len(pdf.pages),
                "file_name": os.path.basename(pdf_path),
                "method": "pdfplumber"
            }

            for page_num, page in enumerate(pdf.pages, 1):
                # Ekstrak teks
                text = page.extract_text()
                if text:
                    result["pages"].append({
                        "page_number": page_num,
                        "text": text,
                        "word_count": len(text.split())
                    })

                # Ekstrak tabel
                tables = page.extract_tables()
                if tables:
                    for table_idx, table in enumerate(tables):
                        result["tables"].append({
                            "page_number": page_num,
                            "table_index": table_idx,
                            "data": table
                        })

        return result

    def _extract_with_pymupdf(self, pdf_path: str) -> Dict:
        """Ekstraksi menggunakan PyMuPDF (cepat dan akurat)"""
        result = {"pages": [], "metadata": {}, "tables": []}

        doc = fitz.open(pdf_path)
        result["metadata"] = {
            "total_pages": len(doc),
            "file_name": os.path.basename(pdf_path),
            "method": "pymupdf"
        }

        for page_num in range(len(doc)):
            page = doc[page_num]
            text = page.get_text()

            if text:
                result["pages"].append({
                    "page_number": page_num + 1,
                    "text": text,
                    "word_count": len(text.split())
                })

        doc.close()
        return result

    def _extract_with_pdfminer(self, pdf_path: str) -> Dict:
        """Ekstraksi menggunakan pdfminer (bagus untuk layout kompleks)"""
        result = {"pages": [], "metadata": {}, "tables": []}

        # Ekstrak seluruh teks
        text = extract_text(pdf_path, laparams=LAParams())

        # Split berdasarkan page breaks (estimasi)
        pages = text.split('\f')  # Form feed character

        result["metadata"] = {
            "total_pages": len(pages),
            "file_name": os.path.basename(pdf_path),
            "method": "pdfminer"
        }

        for page_num, page_text in enumerate(pages, 1):
            if page_text.strip():
                result["pages"].append({
                    "page_number": page_num,
                    "text": page_text,
                    "word_count": len(page_text.split())
                })

        return result

    def classify_page_content(self, page_text: str) -> Dict:
        """
        Klasifikasi konten halaman berdasarkan financial keywords
        """
        text_lower = page_text.lower()

        # Hitung skor berdasarkan keywords
        financial_score = sum(1 for keyword in self.financial_keywords
                            if keyword in text_lower)

        # Deteksi jenis laporan
        report_type = "unknown"
        if any(word in text_lower for word in ["neraca", "balance sheet"]):
            report_type = "balance_sheet"
        elif any(word in text_lower for word in ["laba rugi", "income statement"]):
            report_type = "income_statement"
        elif any(word in text_lower for word in ["arus kas", "cash flow"]):
            report_type = "cash_flow"
        elif any(word in text_lower for word in ["catatan", "notes"]):
            report_type = "notes"

        # Deteksi bahasa
        try:
            language = detect(page_text)
        except:
            language = "unknown"

        return {
            "financial_score": financial_score,
            "report_type": report_type,
            "language": language,
            "is_financial": financial_score > 3,
            "readability_score": textstat.flesch_reading_ease(page_text) if language == 'en' else 0
        }

    def clean_and_preprocess_text(self, text: str) -> str:
        """
        Membersihkan dan preprocessing teks
        """
        # Hapus karakter yang tidak perlu
        text = re.sub(r'\s+', ' ', text)  # Multiple spaces to single
        text = re.sub(r'\n+', '\n', text)  # Multiple newlines to single
        text = re.sub(r'[^\w\s\.,\-\(\)%]', '', text)  # Keep only alphanumeric, spaces, and basic punctuation

        # Hapus nomor halaman
        text = re.sub(r'(?i)halaman\s+\d+', '', text)
        text = re.sub(r'(?i)page\s+\d+', '', text)

        # Normalisasi angka
        text = re.sub(r'(\d+)\.(\d+)\.(\d+)', r'\1\2\3', text)  # Remove dots in numbers

        return text.strip()

    def extract_financial_sections(self, pages: List[Dict]) -> List[Dict]:
        """
        Ekstraksi bagian-bagian penting dari laporan keuangan
        """
        sections = []

        for page in pages:
            page_text = page["text"]
            page_num = page["page_number"]

            # Klasifikasi konten
            classification = self.classify_page_content(page_text)

            # Hanya proses halaman yang relevan secara finansial
            if classification["is_financial"]:
                # Bersihkan teks
                cleaned_text = self.clean_and_preprocess_text(page_text)

                # Pecah menjadi paragraf
                paragraphs = [p.strip() for p in cleaned_text.split('\n')
                            if p.strip() and len(p.strip()) > 50]

                for para_idx, paragraph in enumerate(paragraphs):
                    if len(paragraph) > 100:  # Minimal length
                        sections.append({
                            "page_number": page_num,
                            "paragraph_index": para_idx,
                            "text": paragraph,
                            "report_type": classification["report_type"],
                            "financial_score": classification["financial_score"],
                            "word_count": len(paragraph.split())
                        })

        return sections

    def create_summary_pairs(self, sections: List[Dict]) -> List[Dict]:
        """
        Membuat pasangan teks-summary untuk training
        """
        pairs = []

        for section in sections:
            text = section["text"]

            # Skip jika teks terlalu pendek atau panjang
            if len(text) < 200 or len(text) > 2000:
                continue

            # Generate summary menggunakan rule-based approach
            summary = self._generate_rule_based_summary(text, section["report_type"])

            if summary:
                pairs.append({
                    "text": text,
                    "summary": summary,
                    "metadata": {
                        "page_number": section["page_number"],
                        "report_type": section["report_type"],
                        "financial_score": section["financial_score"],
                        "word_count": section["word_count"]
                    }
                })

        return pairs

    def _generate_rule_based_summary(self, text: str, report_type: str) -> str:
        """
        Generate summary menggunakan rule-based approach
        """
        # Ekstrak angka dan metrics penting
        numbers = re.findall(r'Rp\s*[\d,.]+ (?:triliun|miliar|juta)', text)
        percentages = re.findall(r'\d+(?:\.\d+)?%', text)

        # Ekstrak kalimat kunci berdasarkan keywords
        sentences = re.split(r'[.!?]', text)
        key_sentences = []

        for sentence in sentences:
            sentence = sentence.strip()
            if len(sentence) > 20:
                # Hitung skor kalimat
                score = sum(1 for keyword in self.financial_keywords
                           if keyword.lower() in sentence.lower())

                if score > 0:
                    key_sentences.append((sentence, score))

        # Ambil kalimat dengan skor tertinggi
        key_sentences.sort(key=lambda x: x[1], reverse=True)

        # Buat summary
        summary_parts = []

        # Tambahkan informasi numerik
        if numbers:
            summary_parts.extend(numbers[:2])  # Max 2 numbers

        # Tambahkan kalimat kunci
        if key_sentences:
            summary_parts.append(key_sentences[0][0][:100])  # Truncate if too long

        if summary_parts:
            return '. '.join(summary_parts)

        return ""

    def process_pdf_folder(self, folder_path: str, output_path: str = None) -> List[Dict]:
        """
        Proses seluruh folder berisi PDF laporan keuangan
        """
        pdf_files = []

        # Cari semua file PDF
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                if any(file.endswith(ext) for ext in self.supported_formats):
                    pdf_files.append(os.path.join(root, file))

        print(f"📁 Found {len(pdf_files)} PDF files")

        all_training_pairs = []
        processing_stats = []

        for pdf_path in pdf_files:
            print(f"📄 Processing: {os.path.basename(pdf_path)}")

            # Ekstraksi teks
            extracted_data = self.extract_text_from_pdf(pdf_path)

            if extracted_data["pages"]:
                # Ekstraksi section finansial
                sections = self.extract_financial_sections(extracted_data["pages"])

                # Buat training pairs
                pairs = self.create_summary_pairs(sections)

                all_training_pairs.extend(pairs)

                # Simpan statistik
                processing_stats.append({
                    "file_name": os.path.basename(pdf_path),
                    "total_pages": extracted_data["metadata"]["total_pages"],
                    "financial_sections": len(sections),
                    "training_pairs": len(pairs)
                })

                print(f"✅ Extracted {len(pairs)} training pairs")
            else:
                print(f"⚠️ No text extracted from {pdf_path}")

        # Simpan hasil jika diperlukan
        if output_path:
            with open(output_path, 'w', encoding='utf-8') as f:
                json.dump(all_training_pairs, f, ensure_ascii=False, indent=2)
            print(f"💾 Training data saved to {output_path}")

        # Print statistik
        print("\n📊 Processing Statistics:")
        stats_df = pd.DataFrame(processing_stats)
        print(stats_df.to_string(index=False))

        print(f"\n🎯 Total training pairs: {len(all_training_pairs)}")

        return all_training_pairs

# =============================================================================
# 4. ENHANCED FINANCIAL SUMMARIZER
# =============================================================================

class EnhancedFinancialSummarizer:
    """
    Enhanced version dengan support untuk PDF processing
    """

    def __init__(self, model_name: str = "t5-small"):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model_name = model_name
        self.tokenizer = None
        self.model = None
        self.pdf_extractor = PDFFinancialExtractor()

        print(f"🔧 Using device: {self.device}")
        print(f"🤖 Model: {self.model_name}")

    def setup_model(self):
        """Setup model"""
        try:
            print("🤖 Loading model...")
            self.tokenizer = T5Tokenizer.from_pretrained(self.model_name)
            self.model = T5ForConditionalGeneration.from_pretrained(self.model_name)
            self.model = self.model.to(self.device)
            print("✅ Model loaded successfully")
            return True
        except Exception as e:
            print(f"❌ Error loading model: {e}")
            return False

    def prepare_training_data_from_pdfs(self, pdf_folder_path: str) -> Tuple[List[str], List[str]]:
        """
        Prepare training data dari folder PDF
        """
        print("📁 Processing PDF files...")

        # Ekstraksi data dari PDF
        training_pairs = self.pdf_extractor.process_pdf_folder(pdf_folder_path)

        if not training_pairs:
            print("❌ No training data extracted from PDFs")
            return [], []

        # Konversi ke format yang diperlukan
        texts = []
        summaries = []

        for pair in training_pairs:
            # Tambahkan prefix untuk T5
            text = f"summarize: {pair['text']}"
            texts.append(text)
            summaries.append(pair['summary'])

        print(f"✅ Prepared {len(texts)} training samples")
        return texts, summaries

    def train_with_pdf_data(self, pdf_folder_path: str, output_dir: str = './pdf_financial_model'):
        """
        Training dengan data PDF
        """
        print("🚀 Starting PDF-based training...")

        # Setup model
        if not self.setup_model():
            return None

        # Prepare data
        texts, summaries = self.prepare_training_data_from_pdfs(pdf_folder_path)

        if not texts:
            print("❌ No training data available")
            return None

        # Split data
        train_texts, val_texts, train_summaries, val_summaries = train_test_split(
            texts, summaries, test_size=0.2, random_state=42
        )

        # Create datasets
        train_dataset = Dataset.from_dict({
            'input_text': train_texts,
            'target_text': train_summaries
        })

        val_dataset = Dataset.from_dict({
            'input_text': val_texts,
            'target_text': val_summaries
        })

        # Tokenize
        def tokenize_function(examples):
            model_inputs = self.tokenizer(
                examples['input_text'],
                max_length=512,
                truncation=True,
                padding=True
            )

            with self.tokenizer.as_target_tokenizer():
                labels = self.tokenizer(
                    examples['target_text'],
                    max_length=128,
                    truncation=True,
                    padding=True
                )

            model_inputs["labels"] = labels["input_ids"]
            return model_inputs

        train_tokenized = train_dataset.map(tokenize_function, batched=True)
        val_tokenized = val_dataset.map(tokenize_function, batched=True)

        # Training arguments
        training_args = TrainingArguments(
            output_dir=output_dir,
            num_train_epochs=3,
            per_device_train_batch_size=2,
            per_device_eval_batch_size=2,
            warmup_steps=500,
            weight_decay=0.01,
            logging_dir=f'{output_dir}/logs',
            logging_steps=10,
            evaluation_strategy="steps",
            eval_steps=50,
            save_strategy="steps",
            save_steps=100,
            load_best_model_at_end=True,
            dataloader_pin_memory=False,
            report_to=None,
            fp16=True if torch.cuda.is_available() else False,
        )

        # Data collator
        data_collator = DataCollatorForSeq2Seq(
            tokenizer=self.tokenizer,
            model=self.model,
            padding=True
        )

        # Trainer
        trainer = Trainer(
            model=self.model,
            args=training_args,
            train_dataset=train_tokenized,
            eval_dataset=val_tokenized,
            data_collator=data_collator,
            tokenizer=self.tokenizer,
        )

        # Train
        try:
            trainer.train()

            # Save model
            trainer.save_model(output_dir)
            self.tokenizer.save_pretrained(output_dir)

            print("✅ Training completed and model saved!")
            return trainer

        except Exception as e:
            print(f"❌ Training failed: {e}")
            return None

    def summarize_new_pdf(self, pdf_path: str) -> Dict:
        """
        Summarize laporan keuangan PDF baru
        """
        if self.model is None:
            print("❌ Model not loaded")
            return {}

        # Ekstrak teks dari PDF
        extracted_data = self.pdf_extractor.extract_text_from_pdf(pdf_path)

        if not extracted_data["pages"]:
            print("❌ No text extracted from PDF")
            return {}

        # Ekstrak section finansial
        sections = self.pdf_extractor.extract_financial_sections(extracted_data["pages"])

        summaries = []

        for section in sections:
            if len(section["text"]) > 200:  # Minimal length
                # Tokenize
                input_text = f"summarize: {section['text']}"
                inputs = self.tokenizer(
                    input_text,
                    max_length=512,
                    truncation=True,
                    return_tensors="pt"
                ).to(self.device)

                # Generate summary
                with torch.no_grad():
                    outputs = self.model.generate(
                        inputs["input_ids"],
                        max_length=128,
                        num_beams=4,
                        early_stopping=True,
                        temperature=0.7,
                        do_sample=True
                    )

                summary = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

                summaries.append({
                    "page_number": section["page_number"],
                    "report_type": section["report_type"],
                    "original_text": section["text"][:200] + "...",
                    "summary": summary
                })

        return {
            "file_name": os.path.basename(pdf_path),
            "total_pages": extracted_data["metadata"]["total_pages"],
            "summaries": summaries
        }

# =============================================================================
# 5. MAIN EXECUTION FUNCTION
# =============================================================================

def main_pdf_processing():
    """
    Main function untuk processing PDF
    """
    print("🎯 PDF Financial Report Summarization")
    print("="*50)

    # Initialize
    summarizer = EnhancedFinancialSummarizer()

    # Example usage - ganti dengan path folder PDF Anda
    pdf_folder = "/content/drive/MyDrive/DATASET/Eksperimen_TA"  # Ganti dengan path yang sesuai

    # Pastikan folder ada
    if not os.path.exists(pdf_folder):
        print(f"❌ Folder {pdf_folder} tidak ditemukan")
        print("📁 Silakan buat folder dan upload file PDF laporan keuangan")
        return

    # Train model dengan data PDF
    trainer = summarizer.train_with_pdf_data(pdf_folder)

    if trainer:
        print("🎉 Training berhasil!")

        # Test dengan PDF baru
        test_pdf = "/content/drive/MyDrive/test_report.pdf"  # Ganti dengan path test file

        if os.path.exists(test_pdf):
            print(f"🧪 Testing dengan {test_pdf}")
            result = summarizer.summarize_new_pdf(test_pdf)

            if result:
                print(f"\n📊 Hasil Summary untuk {result['file_name']}:")
                for i, summary in enumerate(result['summaries'], 1):
                    print(f"\n{i}. Halaman {summary['page_number']} ({summary['report_type']}):")
                    print(f"   Summary: {summary['summary']}")

# =============================================================================
# 6. USAGE INSTRUCTIONS
# =============================================================================

"""
INSTRUKSI PENGGUNAAN UNTUK PDF MULTI-HALAMAN:

1. PERSIAPAN:
   - Upload file PDF laporan keuangan ke Google Drive
   - Buat folder khusus, misalnya: /content/drive/MyDrive/financial_reports/
   - Pastikan PDF berisi laporan keuangan dalam bahasa Indonesia/Inggris

2. STRUKTUR FOLDER:
   financial_reports/
   ├── company_a_2023.pdf
   ├── company_b_2023.pdf
   ├── company_c_2024.pdf
   └── ...

3. EKSEKUSI:
   - Jalankan install_pdf_packages() terlebih dahulu
   - Restart runtime
   - Jalankan main_pdf_processing()

4. FITUR YANG DIDUKUNG:
   - Ekstraksi multi-halaman PDF
   - Deteksi otomatis section finansial
   - Klasifikasi jenis laporan (neraca, laba rugi, cash flow)
   - Ekstraksi tabel finansial
   - Preprocessing teks otomatis
   - Generate summary berdasarkan rule-based + ML

5. TROUBLESHOOTING:
   - Jika PDF tidak terbaca: coba method ekstraksi berbeda
   - Jika OCR diperlukan: tambahkan pytesseract processing
   - Jika bahasa tidak terdeteksi: manual language setting

6. OPTIMASI:
   - Gunakan GPU runtime untuk training cepat
   - Batch processing untuk many PDFs
   - Adjust max_length sesuai kebutuhan
"""

# Uncomment untuk menjalankan
main_pdf_processing()
```
